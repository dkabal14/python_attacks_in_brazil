```python
import pandas as pd
import requests as rq
import base64

protocol = 'https'
server = 'apis-sa.highbond.com'
org_id = '34645'
page = 30
page_b64 = base64.b64encode(str(page).encode()).decode()
Token = 'preencher'

url = f'{protocol}://{server}/v1/orgs/{org_id}/signoffs'

headers = {
    'Content-type': 'application/vnd.api+json',
    'Authorization': 'Bearer ' + Token
}

parameters = {
    'fields[signoffs]': 'created_at,updated_at,prepared_at,detail_reviewed_at,general_reviewed_at,supplemental_reviewed_at,specialty_reviewed_at,project,target,preparer,detail_reviewer,general_reviewer,supplemental_reviewer,specialty_reviewer,next_reviewer',
    'page[size]': 50,
    'page[number]': page_b64
}

apiResponse = rq.get(url, params=parameters, headers=headers)
df = pd.json_normalize(apiResponse.json()['data'])
display(df)
```
